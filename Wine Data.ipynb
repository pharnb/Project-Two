```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Read in the csv file to a dataframe
data_df = pd.read_csv("Resources/wine_data.csv")
data_df.head()
```

```python
#Create new dataframe which removes unwanted columns
wine_data_df=data_df[['country', 'designation', 'price', 'province', 'title', 'variety']]
wine_data_df
```

```python
#Drop rows with NAN values
wine_data_df=wine_data_df.dropna(how='any')
wine_data_df
```

```python
#wine_data_df.dtypes
#Create new column to convert price float value to currency
wine_data_df['pricing']= wine_data_df['price'].map("${0:.2f}".format)
wine_data_df
```

```python
wine_df=wine_data_df[['country', 'designation', 'province', 'title', 'variety', 'pricing']]
wine_df
```

```python
rds_connection_string = "postgres:stopdying@bootcamp.cbtoq3riq9z0.us-west-1.rds.amazonaws.com:5432/project2_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
wine_df.to_sql(name='wine_data', con=engine, if_exists='append', index=False)
```
