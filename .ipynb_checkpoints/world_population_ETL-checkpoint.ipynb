```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
### Store CSV into DataFrame

csv_file = "Resources/world_pop.csv"
world_pop_df = pd.read_csv(csv_file)
world_pop_df.head()
```

```python
### Create new data with select columns
world_pop_col = ['Country (or dependency)', 'Population (2020)', 'Land Area (Km²)', 'Density (P/Km²)', 'World Share']
new_world_pop_df = world_pop_df[world_pop_col].copy()
new_world_pop_df.head()
```

```python
### Rename columns, set index, add commas to numbers for readability

renamed_world_pop_df = new_world_pop_df.rename(columns={'Country (or dependency)': 'country', 
                                                         'Population (2020)': 'population', 
                                                         'Density (P/Km²)': 'density', 
                                                         'Land Area (Km²)': 'land_area', 
                                                         'World Share': 'world_percentage'})
renamed_world_pop_df.set_index("country", inplace=True)

# renamed_world_pop_df = renamed_world_pop_df.astype('object')#.dtypes
# print(renamed_world_pop_df.dtypes)
renamed_world_pop_df.population = renamed_world_pop_df.population.apply(lambda x : "{:,}".format(x))
renamed_world_pop_df.land_area = renamed_world_pop_df.land_area.apply(lambda x : "{:,}".format(x))

renamed_world_pop_df.head()
# print(renamed_world_pop_df.dtypes)
```
